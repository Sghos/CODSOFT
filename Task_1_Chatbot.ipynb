```python
# STEP 1: Install the website maker tool
!pip install gradio -q

import gradio as gr

# STEP 2: The "Brain" of the Chatbot
def chatbot_response(user_input):
    # This makes all text small letters so the computer isn't confused
    user_input = user_input.lower()

    # Rule 1: Greetings
    if "hello" in user_input or "hi" in user_input:
        return "Chatbot: Hi there! It's so nice to meet you. How can I help you today?"

    # Rule 2: Personal Identity
    elif "who are you" in user_input or "your name" in user_input:
        return "Chatbot: I am a specialized AI assistant built for my CodSoft internship! My job is to learn how to talk to humans."

    # Rule 3: Feelings
    elif "how are you" in user_input:
        return "Chatbot: I'm feeling very digital today! I have plenty of battery and I'm ready to help you. How are YOU feeling?"

    # Rule 4: Weather (The "Human" touch)
    elif "weather" in user_input:
        return "Chatbot: Inside this computer, it is a perfect 72 degrees. I hope it's nice outside where you are!"

    # Rule 5: Explaining AI (Task Requirement)
    elif "what is ai" in user_input:
        return "Chatbot: AI stands for Artificial Intelligence. It is basically giving a computer a set of rules so it can act smart, just like I'm doing right now!"

    # Rule 6: Internship/Task Information
    elif "internship" in user_input or "task" in user_input:
        return "Chatbot: You are doing great! To finish this internship, you need to complete 3 tasks and post a video of them on LinkedIn with #codsoft."

    # Rule 7: Politeness
    elif "thank" in user_input:
        return "Chatbot: You are very welcome! Helping you is what I was programmed for."

    # Rule 8: Saying Goodbye
    elif "bye" in user_input or "goodbye" in user_input:
        return "Chatbot: Goodbye! Don't forget to save your work to GitHub! See you soon."

    # The "Safety Net" - Keeps the conversation going
    else:
        return "Chatbot: That is a very interesting thought! Since I am still learning, I might not know about that yet. Can you tell me more, or ask me something like 'What is AI?'"

# STEP 3: The "Face" of the Chatbot (The Interface)
interface = gr.Interface(
    fn=chatbot_response,
    inputs=gr.Textbox(lines=2, placeholder="Type your message here..."),
    outputs="text",
    title="ðŸ¤– CodSoft Internship: Task 1 Chatbot",
    description="This is a rule-based chatbot built using Python. It uses 'if-else' logic to understand and reply to you!"
)

# STEP 4: Launch!
interface.launch(share=True)
```
