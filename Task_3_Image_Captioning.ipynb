```python
import gradio as gr
from transformers import pipeline

# 1. Load the smart "vision" brain
captioner = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")

def describe_image(image):
    if image is None: return "Please upload an image first!"
    result = captioner(image)
    return result[0]['generated_text'].capitalize()

# 2. Design the "Professional System" Look
# We use 'theme=gr.themes.Soft()' to make it look modern and clean
with gr.Blocks(theme=gr.themes.Soft(), css=".container { max-width: 900px; margin: auto; }") as demo:

    gr.Markdown("# ðŸŽ¨ AI Vision System")
    gr.Markdown("### Professional Image Captioning Tool for CodSoft Internship")

    with gr.Row():
        # Left side: Upload area
        with gr.Column(scale=1):
            input_img = gr.Image(type="pil", label="Step 1: Upload Photo")
            btn = gr.Button("Analyze Image", variant="primary")

        # Right side: Professional Output area
        with gr.Column(scale=1):
            output_text = gr.Textbox(label="Step 2: AI Result", placeholder="The description will appear here...")
            gr.Examples(examples=["https://images.unsplash.com/photo-1514888286974-6c03e2ca1dba"], inputs=input_img)

    btn.click(fn=describe_image, inputs=input_img, outputs=output_text)

# 3. Launch with 'share=True' to get that NEW TAB link
demo.launch(share=True)
```
